```python
# Then the change point is detected by iteratively computing the AIC values
import pandas as pd
import numpy as np
# Reference date and compute days_from_t0
t0 = 0

# run Fortran program and get AIC
def read_work_etas(filepath):
    return pd.read_csv(filepath, delim_whitespace=True, header=None, names=[
        'seq_num', 'longitude', 'latitude', 'mag', 'days_from_t0', 'depth', 'year', 'month', 'day'
    ])

# write subset data to work.etas file
def write_work_etas(subset_data, filename):
    subset_data.to_csv(filename, sep=' ', header=False, index=False)

# extract AIC value from the Fortran output
def extract_aic(fortran_output):
    import re
    match = re.search(r'aic/2\s*=\s*([-0-9.eE+]+)', fortran_output)
    if match:
        aic_half_value = float(match.group(1))
        return 2 * aic_half_value
    else:
        raise ValueError("AIC value not found in Fortran output")
# run Fortran program and get AIC
def run_fortran_and_get_aic(input_file, control_file):
    # Call the Fortran program and capture the output
    fortran_output = !./etas
    fortran_output = "\n".join(fortran_output)

    # Extract AIC value from the Fortran output
    aic_value = extract_aic(fortran_output)
    return aic_value
```

```python
def write_etas_open(filename, zts, zte, tarst):
    content = f"""9 2
{zts:.3f} {zte:.3f} {tarst:.2f}
3.6 7.2
0.34378E+00 0.24709E+00 0.49796E-02 0.63649E+00 0.11189E+01
{zts:.3f} {zte:.3f} {tarst:.2f}
3.6 7.2
-0.4753056E+04
0.34378E+00 0.24709E+00 0.49796E-02 0.63649E+00 0.11189E+01
"""
    with open(filename, 'w') as f:
        f.write(content)

import csv

aic_filename = 'combined_aic_values.csv'
with open(aic_filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Changepoint', 'Combined_AIC'])
```

```python
data = read_work_etas('/content/work.etas')

# Define input and output file paths
control_file = 'etas.open'

combined_aic_values = []
# Define a list of possible changepoints
changepoints = np.unique(data['days_from_t0'])
```

```python
# Loop over possible changepoints
for i, changepoint in enumerate(changepoints):
    # Split the data into two parts
    data1 = data[data['days_from_t0'] < changepoint]
    data2 = data[data['days_from_t0'] >= changepoint]

    # Define filenames for the two parts
    input_file1 = f'work_part1_{i}.etas'
    input_file2 = f'work_part2_{i}.etas'

    # Write the two parts to separate files and run Fortran
    write_work_etas(data1, input_file1)
    write_etas_open(control_file, zts=0.0, zte=changepoint, tarst=5.01)
    aic1 = run_fortran_and_get_aic(input_file1, control_file)

    write_work_etas(data2, input_file2)
    write_etas_open(control_file, zts=changepoint, zte=data['days_from_t0'].max(), tarst=changepoint+0.01)
    aic2 = run_fortran_and_get_aic(input_file2, control_file)

    # Combine the AIC values
    combined_aic = aic1 + aic2
    combined_aic_values.append(combined_aic)
    with open(aic_filename, mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([changepoint, combined_aic])
```

```python
import matplotlib.pyplot as plt
optimal_changepoint = changepoints[np.argmin(combined_aic_values)]

# Plot the combined AIC values over changepoints
plt.plot(changepoints, combined_aic_values, marker='o')
plt.axvline(optimal_changepoint, color='red', linestyle='--')
plt.xlabel('Changepoint')
plt.ylabel('Combined AIC Value')
plt.title('Combined AIC over Changepoints')
plt.show()

# Output the optimal changepoint
print("The optimal changepoint is at position:", optimal_changepoint)
files.download(aic_filename)
```
